# 1. 퍼셉트론이란

퍼셉트론은 다수의 신호를 입력받아 하나의 신호를 출력한다. 아래 그림과 같이 입력 신호가 뉴런에 보내질 때 각각 고유한 가중치가 곱해진다. 뉴런에서 보내온 신호의 총합이 정해진 한계를 넘을 때만 1을 출력한다. 



# 2. 단순한 논리회로

 - AND 게이트
 
 

```python
# AND 게이트 구현
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp >= theta:
        return 1
```

```python
AND(0,0.5) # 1
AND(0.5,0.5) #0
AND(1,1) # 1
```

```python
# 가중치와 편향 도입
import numpy as np
x = np.array([0,1]) # 입력
w = np.array([0.5,0.5]) # 가중치
b = -0.7 # 편향
w*x # array([0. , 0.5])

np.sum(w*x) # 0.5
np.sum(w*x) + b # -0.199999999
```

```python
# 가중치와 편향 구현하기
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5,0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
AND(1,0.5)
```

- OR 게이트



```python
# OR 게이트 구현
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1


OR(1,0) # 1
OR(0.1,0.3) # 0
```

 - NAND 게이트
    


```python
# NAND 게이트 구현

def NAND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([-0.5, -0.5]) # Not AND
    b = 0.7
    tmp = sum(w*x) + b
    
    if tmp <= 0:
        return 0
    else:
        return 1
    
NAND(0,1) # 1
NAND(1,1) # 0
```

# 3 퍼셉트론의 한계

 - XOR 게이트

 XOR 게이트는 배타적 논리합이라는 논리회로써 x1과 x2 중 한쪽이 1일 때만 1을 출력한다. XOR의 구현이 안되는 이유는 비선형 구조를 띄기 때문에 일반 퍼셉트론으로 구현할 수 없다.
 
 

- 다층 퍼셉트론으로 XOR를 표현하기





```python
# XOR 구현하기

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

XOR(0,0) # 0
XOR(1,0) # 1
XOR(0,1) # 1
XOR(1,1)
```



이와 같이 단층 퍼셉트론으로 표현하지 못한 것을 층을 하나 늘려서 구현할 수 있다.

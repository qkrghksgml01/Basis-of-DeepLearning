 - AND 
 
 

```python
# AND  
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp >= theta:
        return 1
```

```python
AND(0,0.5) # 1
AND(0.5,0.5) #0
AND(1,1) # 1
```

```python
#   
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5,0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
AND(1,0.5)
```

- OR 



```python
# OR  
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1


OR(1,0) # 1
OR(0.1,0.3) # 0
```

 - NAND 
    


```python
# NAND  

def NAND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([-0.5, -0.5]) # Not AND
    b = 0.7
    tmp = sum(w*x) + b
    
    if tmp <= 0:
        return 0
    else:
        return 1
    
NAND(0,1) # 1
NAND(1,1) # 0
```

-   XOR 





```python
# XOR 

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

XOR(0,0) # 0
XOR(1,0) # 1
XOR(0,1) # 1
XOR(1,1)
```

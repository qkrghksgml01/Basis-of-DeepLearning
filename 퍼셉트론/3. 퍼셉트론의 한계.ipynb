 - XOR 게이트

 XOR 게이트는 배타적 논리합이라는 논리회로써 x1과 x2 중 한쪽이 1일 때만 1을 출력한다. XOR의 구현이 안되는 이유는 비선형 구조를 띄기 때문에 일반 퍼셉트론으로 구현할 수 없다.
 


- 다층 퍼셉트론으로 XOR를 표현하기





```python
# XOR 구현하기
from functions import NAND, OR, AND # 2장 과정

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

XOR(0,0) # 0
XOR(1,0) # 1
XOR(0,1) # 1
XOR(1,1)
```



이와 같이 단층 퍼셉트론으로 표현하지 못한 것을 층을 하나 늘려서 구현할 수 있다.
